```python
cc1 = list(combinations(list(range(len(sec_returns2.columns))), num_stocks_in_portfolio))
assert len(cc1) == len(cc)
```

```python
_GLOBAL_ = list(range(len(cc)))
_VOLATILES_ = list(range(len(cc)))
_DS_ = sec_returns2
_M_ = _DS_.to_numpy()
combinations_str = cc
combinations_int = cc1
weights1 = np.array([0.5, 0.5, 0.5])
```

```python
def Calculate_Loc(_id, _from, _to):
    
    print(f'Start calculation with loc from: {_from} to {_to}')

    for i in range(_from, _to):

        temp = _DS_.loc[:,combinations_str[i]]
        
        cov_matrix = (temp.cov()*252).to_numpy()
        temp_vol = np.dot(weights1.T, np.dot(cov_matrix, weights1))**.5
        _VOLATILES_[i] = temp_vol

        ret = np.dot(weights1, temp.mean()*252)
        _GLOBAL_[i] = ret

        del temp
```

```python
def Calculate_NP(_id, _from, _to):
    
    print(f'Start {_id} calculation with numpy from: {_from} to {_to}')

    for i in range(_from, _to):

        temp = _M_[:, combinations_int[i]]

        cov_matrix = (np.cov(temp.T)*252)
        temp_vol = np.dot(weights1.T, np.dot(cov_matrix, weights1))**.5
        _VOLATILES_[i] = temp_vol
        
        ret = np.dot(weights1, temp.mean()*252)
        _GLOBAL_[i] = ret
        
        del temp
    
    print(f'Stop {_id} calculation with numpy from: {_from} to {_to}')
    return
```

```python
STEP = 20000
TOTAL = len(cc1)
limits = list(range(0, TOTAL+1, STEP))[1:]
threads = []
cur = 0
thread_id = 0

for l in limits:
    thread_id+=1
    x = threading.Thread(target=Calculate_NP, args=(thread_id, cur, l))
    x.start()
    threads.append(x)
    cur = l

print(f'Total threads: {len(threads)}')
```

```python
start = time.time()

for t in threads:
    t.join()
    
end = time.time()
print(end - start)
```
